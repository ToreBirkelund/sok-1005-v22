```python
# packages
#import for statsmodels
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from statsmodels.regression.linear_model import OLS

# bokeh pakker
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.plotting import output_file


#Samarbeid: Kenneth Benonisen, Tore Birkelund.
#Kode hentet fra Espen Sirnes.
```

```python
# henter data fra https://ourworldindata.org/explorers/natural-resources?facet=none&Resource=Oil&Metric=Reserves&Count=Total&country=USA~RUS~NOR~GBR~DEU~SWE~CHN

df = pd.read_csv('natural-resources.csv', index_col=0)
df
```

```python
df_filtered = df[df['Entity'].isin(["Norway", "Canada", "Russia", "United States", "Nigeria", "Venezuela", "Ukraine", "Denmark"])]
df_filtered
```

```python
df_filtered_2020 = df_filtered[df_filtered["Year"] == 2020]
df_filtered_2020
```

```python
# creating a dictionary of continents with norwegian translation:
continents={'Venezuela':'Venezuela', 'United States':'USA', 'Ukraine':'Ukraina', 'Russia':'Russland', 'Norway':'Norge', 'Nigeria':'Nigeria', 'Denmark':'Danmark', 'Canada':'Canada'}

#assigning a color hex-code to each continent
colors={'Venezuela': '#E14827', 'United States': '#84E127', 'Ukraine': '#2792E1', 'Russia': '#E04A6C', 'Norway' : '#DEB887', 'Nigeria': 'darkgreen', 'Denmark': 'purple', 'Canada': '#5F9EA0'}

df_filtered_2020['Entity_no'] = df_filtered_2020['Entity'].apply(lambda cont:  continents[cont])
df_filtered_2020['colors'] = df_filtered_2020['Entity'].apply(lambda cont: colors[cont])
df_filtered_2020
```

```python
df_filtered_2020['size'] = df_filtered_2020['Population'].apply(lambda pop: pop**0.5/300)
df_filtered_2020
```

```python
df_filtered_2020['Oil_production_rnd'] = df_filtered_2020['Oil production'].apply(lambda l: int(l))
df_filtered_2020['Oil_reserves_per_capita_rnd'] = df_filtered_2020['Oil reserves per capita'].apply(lambda l: int(l))
df_filtered_2020['pop_mill_str'] = df_filtered_2020['Population'].apply(lambda pop:  '{:,}'.format(int(pop/1000000)))
df_filtered_2020
```

```python
#creating figure:
p = figure(
        title = "Olje reserver, produksjon og land", 
        x_axis_label = 'Olje reserver per capita',
        y_axis_label = 'Oil production',
    
        x_axis_type = "log",
        y_axis_type = "log",
    
        tools="hover", 
        tooltips = [
            ("Land", "@Entity_no"),
            ("Olje reserver", "@Oil_reserves_per_capita_rnd"),
            ("Befolkning", "@pop_mill_str"),
            ("Olje Produksjon", "@Oil_production_rnd")
            ],
    
        plot_height = 580,
        plot_width = 980)
```

```python
p.scatter(
    source=df_filtered_2020,
    x= 'Oil reserves per capita', 
    y= 'Oil production',
    
    size='size', 
    color= 'colors',
    legend_field='Entity_no',
    alpha= 0.8
    )

#formatting:
p.xaxis.major_label_overrides = { 1000: '1k', 10000: '10k', 100000: '100k' }
p.legend.location = "top_left"

p.legend[0].border_line_alpha=0
p.outline_line_alpha=0
p.grid[0].grid_line_alpha=0
p.grid[1].grid_line_alpha=0


#Creating the graph and saving as html
output_notebook()
output_file("Olje_reserver.html")
show(p)
```
